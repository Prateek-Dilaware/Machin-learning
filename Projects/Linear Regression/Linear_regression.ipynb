```python
import matplotlib.pyplot as plt

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
visits = [120, 135, 150, 170, 160]

plt.plot(days, visits, marker='o')
plt.title("Website Visits Over a Week")
plt.xlabel("Day")
plt.ylabel("Visits")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import pandas as pd

data = pd.DataFrame({
    'Department': ['HR', 'Tech', 'Finance', 'Sales'],
    'Salary': [50000, 85000, 70000, 60000]
})

sns.barplot(x='Department', y='Salary', data=data)
```
